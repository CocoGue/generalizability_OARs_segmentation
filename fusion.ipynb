```python
from SimpleITK import GetArrayFromImage, GetImageFromArray, ReadImage, WriteImage
import os
import numpy as np

#We have one folder with the segmentation maps of each organ as we are fusing them.
datapath_aorta = "result_Aorta_Fusion"
list_files_aorta = sorted(os.listdir(datapath_aorta))

datapath_heart = "result_Heart_Fusion"
list_files_heart = sorted(os.listdir(datapath_heart))

datapath_oeso = "result_Oesophagus_Fusion"
list_files_oeso = sorted(os.listdir(datapath_oeso))

datapath_trachea = "result_Trachea_Fusion"
list_files_trachea = sorted(os.listdir(datapath_trachea))

#Every patient should have the same filename inside each organ's folder
for file in list_files_aorta: 
    patient = file
    print(patient)
    
    image_path_aorta = os.path.join(datapath_aorta, patient)
    image_path_heart = os.path.join(datapath_heart, patient)
    image_path_oeso = os.path.join(datapath_oeso, patient)
    image_path_trachea = os.path.join(datapath_trachea, patient)

    ct_aorta = ReadImage(image_path_aorta)
    ct_heart = ReadImage(image_path_heart)
    ct_oeso = ReadImage(image_path_oeso)
    ct_trachea = ReadImage(image_path_trachea)

    ct_aorta_np=GetArrayFromImage(ct_aorta)
    ct_heart_np=GetArrayFromImage(ct_heart)
    ct_oeso_np=GetArrayFromImage(ct_oeso)
    ct_trachea_np=GetArrayFromImage(ct_trachea)
    
    #Esophagus should have label 1
    #Heart should have label 2
    #Trachea should have label 3
    #Aorta should have label 4

    #To be sure we have only voxels representing aorta
    ct_aorta_np[ct_aorta_np == 4] = 0
    ct_aorta_np[ct_aorta_np == 2] = 0
    ct_aorta_np[ct_aorta_np == 3] = 0
    ct_aorta_np[ct_aorta_np == 1] = 4
    
    ct_aorta_np[ct_heart_np == 1] = 2
    ct_aorta_np[ct_oeso_np == 1] = 1
    ct_aorta_np[ct_trachea_np == 1] = 3

    #Step 6: Save new images
    image_to_save = GetImageFromArray(ct_aorta_np)


    image_file_name = "/yourFolderName/result/"+patient
    
    image_to_save.CopyInformation(ct_oeso)
    
    WriteImage(image_to_save, image_file_name)
    print("Image and labelmap saved")
```
